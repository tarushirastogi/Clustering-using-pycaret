```python
from pycaret.utils import version
version()
```

```python
import pandas as pd
from pycaret.clustering import *
```

```python
df = pd.read_csv('/content/Wholesale customers data.csv')
df.head()
```

```python
df.info()
```

**K-Means clustering without Data Pre-Processing**

```python
setup(data = df, verbose=False)
x = create_model('kmeans')
plot_model(x, plot = 'elbow')
```

```python
setup(df,verbose=False)
print("For Cluster 3")
x = create_model('kmeans', num_clusters = 3)
print("For Cluster 4")
x = create_model('kmeans', num_clusters = 4)
print("For Cluster 5")
x = create_model('kmeans', num_clusters = 5)
print("For Cluster 6")
x = create_model('kmeans', num_clusters = 6)
```

```python
setup(df,verbose=False)
kmeansModel = create_model('kmeans', num_clusters = 5)
kmeansPrediction = assign_model(kmeansModel)
kmeansPrediction
```

```python
plot_model(kmeansModel, plot='tsne')
```

**K-Means Clustering with Normalization**

```python
setup(data = df, normalize = True, normalize_method = 'zscore', verbose=False)
x = create_model('kmeans')
plot_model(x, plot = 'elbow')
```

```python
setup(data = df, normalize = True, normalize_method = 'zscore', verbose=False)

print("For Cluster = 3")
x = create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
x = create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
x = create_model('kmeans', num_clusters = 5)

print("For Cluster = 6")
x = create_model('kmeans', num_clusters = 6)
```

```python
setup(data = df, normalize = True, normalize_method = 'zscore', verbose=False)
kmeansModel1 = create_model('kmeans', num_clusters = 5)
plot_model(kmeansModel1, plot = 'tsne')
```

```python
kmeansPrediction1 = assign_model(kmeansModel1)
kmeansPrediction1
```
